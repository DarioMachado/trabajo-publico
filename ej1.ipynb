```python
import sqlite3
import pandas as pd

db_name = "c_sharp_db_15.sqlite"

conn = sqlite3.connect(db_name)
c = conn.cursor()

c.execute("SELECT r.Url, r.Stars, r.Forks, r.ProcessedTime || '  ' || s.ProcessedTime FROM Repo r JOIN Solution s ON r.Id = s.RepoId WHERE r.Stars > 50 AND r.Forks > 100")

print("-------------")


df_select5 = pd.DataFrame(c.fetchall(),columns=['Url', 'Stars', 'Forks', 'Processed Time'])


display(df_select5)
```
